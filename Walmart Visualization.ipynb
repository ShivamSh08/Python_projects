# Calculated fields

```python
sales['Date']=pd.to_datetime(sales['Date'])
sales['Year']=sales['Date'].dt.year
sales['Month']=sales['Date'].dt.month
sales['Day']=sales['Date'].dt.day
sales['MonthYear_value'] = sales['Date'].dt.to_period('M')
sales['Quarter_value'] = sales['Date'].dt.to_period('Q')
sales.head()
```

# Impact of unemployment on Sales

```python
print("The correlation coefficient  between Weekly_Sales and Holiday_Flag is :",
      np.corrcoef(sales['Weekly_Sales'],sales['Unemployment'])[0,1])
```

```python
sales[['Weekly_Sales', 'Unemployment']].head(10)
```

```python
ax1 = sns.jointplot(x='Unemployment',y='Weekly_Sales',data=sales, kind='reg', height = 8)
ax1.fig.suptitle('Relation between Unemployment Rate and Weekly Sales',size = 20)
ax1.fig.subplots_adjust(top=0.95)
ax1.set_axis_labels('Unemployment Rate', 'Weekly Sales in Million Dollars', fontsize=8)
plt.show()
# It seems like the unemployment rate goes high will cause the weekly sales goes down.
```

# Impact of Fuel Price on Sales

```python
ax = sns.jointplot(x='Fuel_Price',y='Weekly_Sales',data=sales, kind='reg', height = 8, color='green')
ax.fig.suptitle('Relation between Fuel Price and Weekly Sales',size = 20)
ax.fig.subplots_adjust(top=0.95)
ax.set_axis_labels('Fuel Price', 'Weekly Sales in Million Dollars', fontsize=8)
plt.show()
# It seems like higher fuel price will not influence weekly sales
```

```python
ax = sns.jointplot(x='Fuel_Price',y='Weekly_Sales', data=sales[sales.Store == 1], kind='reg', height = 8, color='green')
ax.fig.suptitle('Relation between Fuel Price and Weekly Sales',size = 20)
ax.fig.subplots_adjust(top=0.95)
ax.set_axis_labels('Fuel Price', 'Weekly Sales in Million Dollars', fontsize=8)
plt.show()
# It seems like higher fuel price will not influence weekly sales
```

```python
ax = sns.jointplot(x='Temperature',y='Weekly_Sales', data=sales[sales.Store == 33], kind='reg', height = 8, color='green')
ax.fig.suptitle('Relation between Temperature and Weekly Sales',size = 20)
ax.fig.subplots_adjust(top=0.95)
ax.set_axis_labels('Temperature', 'Weekly Sales in Million Dollars', fontsize=8)
plt.show()
# It seems like higher fuel price will not influence weekly sales
```

## Sales by Year month - By Stores

```python
sales_year_store = sales.groupby(['Store', 'MonthYear_value'])['Weekly_Sales'].sum().reset_index()
sales_year_store
```

```python
font_size = 10
fig_size = (14,8)
# pal = "plasma"
plt.figure(figsize=fig_size)

sns.lineplot(x=sales_year_store['MonthYear_value'].astype(str), y='Weekly_Sales', hue='Store', data = sales_year_store)
plt.xticks(size=font_size, rotation=45);
```
